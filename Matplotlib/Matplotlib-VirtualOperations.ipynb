```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpyArray1=np.linspace(0,10,20)
```

```python
numpyArray1
```

```python
numpyArray2=numpyArray1**2
```

```python
numpyArray2
```

```python
(myFigure,myAxis)=plt.subplots()
myAxis.plot(numpyArray1,numpyArray2,color="#AB2276",alpha=0.3) # alpha tranparancy
myAxis.plot(numpyArray2,numpyArray1,color="#6693A8")
plt.show()
```

```python
(newwFigure,newAxis)=plt.subplots()
newAxis.plot(numpyArray1,numpyArray1+2,color="blue",linewidth=2.0,linestyle=":")
newAxis.plot(numpyArray1,numpyArray1+3,color="yellow",linewidth=5.0,linestyle="-.")
newAxis.plot(numpyArray1,numpyArray1+4,color="#669345",linestyle="--") ## : , -.
newAxis.plot(numpyArray1,numpyArray1+5,color="#669345",linestyle="--",marker="+",markerfacecolor="red",markersize=14) ## : , -.
newAxis.plot(numpyArray1,numpyArray1+6,color="#000000",linestyle="--",marker="o",markerfacecolor="white",markersize=4) ## : , -.

plt.show()
```

## Scatter

```python
plt.scatter(numpyArray1,numpyArray2)
plt.show()
```
